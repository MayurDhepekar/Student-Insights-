# QUE.1) HOW MANY UNIQUE STUDENTS ARE INCLUDED IN THE DATASET?

```python
# Count the number of unique students based on their email IDs or any unique identifier
num_unique_students = df['Email ID'].nunique()
print(f"Number of unique students in the dataset: {num_unique_students}")
```

# QUE.2) WHAT IS THE AVERAGE GPA OF THE STUDENT?

```python
# Calculate the average CGPA of the students
average_cgpa = df['CGPA'].mean()
print(f"The average CGPA of the students is: {average_cgpa}")
```

# QUE.3) WHAT IS THE DISTRIBUTION OF STUDENTS ACROSS DIFFERENT GRADUATION YEARS?

```python
# Plot the distribution of students across different graduation years
plt.figure(figsize=(10, 6))
df['Year of Graduation'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('Year of Graduation')
plt.ylabel('Number of Students')
plt.title('Distribution of Students Across Graduation Years')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# QUE.4) WHAT IS THE DISTRIBUTION OF STUDENTS EXPERIENCE WITH PYTHON PROGRAMMING?

```python
# Plot the distribution of students' experience with Python
plt.figure(figsize=(10, 6))
df['Experience with python (Months)'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('Experience with Python (Months)')
plt.ylabel('Number of Students')
plt.title('Distribution of Students\' Experience with Python')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# QUE.5) WHAT IS THE AVERAGE FAMILY INCOME OF THE STUDENT?

```python
# Preprocess the 'Family Income' column to extract numeric values
# Convert 'Lakh+' to numeric values (e.g., 7 Lakh+ will become 700000)
df['Family Income'] = df['Family Income'].apply(lambda x: int(x.split(' ')[0]) * 100000 if 'Lakh+' in x else x)

# Remove non-numeric characters and convert the column to numeric
df['Family Income'] = pd.to_numeric(df['Family Income'].str.replace('[^\d]', ''), errors='coerce')

# Calculate the average family income
average_family_income = df['Family Income'].mean()

# Print the result
print(f'The average family income of the students is: {average_family_income}')
```

# QUE.6) HOW DOES THE CGPA VARY AMONG DIFFERENT COLLEGES? (Show top 5 results only)

```python
# Calculate the average CGPA for each college
college_avg_cgpa = df.groupby('College Name')['CGPA'].mean().sort_values(ascending=False).head(5)

# Plot the average CGPA for top 5 colleges
plt.figure(figsize=(10, 6))
college_avg_cgpa.plot(kind='bar')
plt.xlabel('College Name')
plt.ylabel('Average CGPA')
plt.title('Average CGPA Among Top 5 Colleges')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# QUE.7) ARE THERE ANY OUTLIERS IN THE QUANTITY (number of courses completed) ATTRIBUTE?


```python
# Plot a boxplot to identify outliers in the 'Quantity' attribute
plt.figure(figsize=(6, 4))
plt.boxplot(df['Quantity'], vert=False)
plt.xlabel('Quantity')
plt.title('Box Plot of Quantity (Number of Courses Completed)')
plt.show()
```

```python
# Calculate Interquartile Range (IQR) to identify outliers
Q1 = df['Quantity'].quantile(0.25)
Q3 = df['Quantity'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
# Identify outliers
outliers = df[(df['Quantity'] < lower_bound) | (df['Quantity'] > upper_bound)]
print("Outliers in the 'Quantity' attribute:")
print(outliers)
```

# QUE.8) WHAT IS THE AVERAGE GPA FOR STUDENTS FROM EACH CITY?

```python
# Calculate the average CGPA for students from each city
city_avg_cgpa = df.groupby('City')['CGPA'].mean()

# Print the average CGPA for each city
for city, avg_cgpa in city_avg_cgpa.items():
    print(f"Average CGPA in {city}: {avg_cgpa}")
```

# QUE.13) DO STUDENTS IN LEADERSHIP POSITION DURING THEIR COLLEGE YEARS TEND TO HAVE HIGHER CGPAS OR BETTER EXPECTED SALARY?

```python
# Calculate the average CGPA and expected salary for students with and without leadership positions
avg_cgpa_with_leadership = df[df['Leadership- skills'] == 'yes']['CGPA'].mean()
avg_cgpa_without_leadership = df[df['Leadership- skills'] == 'no']['CGPA'].mean()

avg_salary_with_leadership = df[df['Leadership- skills'] == 'yes']['Expected salary (Lac)'].mean()
avg_salary_without_leadership = df[df['Leadership- skills'] == 'no']['Expected salary (Lac)'].mean()
```

```python
# Print the results
print(f"Average CGPA with leadership: {avg_cgpa_with_leadership}")
print(f"Average CGPA without leadership: {avg_cgpa_without_leadership}")

print(f"Average Expected salary with leadership: {avg_salary_with_leadership}")
print(f"Average Expected salary without leadership: {avg_salary_without_leadership}")
```

# QUE.14) HOW MANY STUDENTS ARE GRADUATING BY THE END OF 2024?

```python
# Convert 'Year of Graduation' to datetime format
df['Year of Graduation'] = pd.to_datetime(df['Year of Graduation'], errors='coerce')

# Filter students graduating by the end of 2024
graduating_by_2024 = df[df['Year of Graduation'].dt.year <= 2024]

# Count the number of students graduating by the end of 2024
num_students_graduating_by_2024 = len(graduating_by_2024)

# Print the result
print(f"Number of students graduating by the end of 2024: {num_students_graduating_by_2024}")
```
